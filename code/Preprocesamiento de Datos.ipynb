#### 2.- Inspección de los datos

```python
# Leemos el archivo json 
import json

with open('US_category_id.json','r') as f:
    data_json=json.loads(f.read())
catDF_US = pd.json_normalize(data_json,record_path=['items'])
catDF_US
```

```python
## A nosotros solo nos interesan 2 columnas de este dataframe generado a partir del json
## "id" y "snippet.title", ya que es lo que vamos a utilizar para nuestro DF principal

catDF_US = catDF_US.drop(['kind','etag','snippet.channelId','snippet.assignable'],axis=1, errors='ignore')
catDF_US.info()
```

```python
## Podemos observar que id es de tipo object, hay que convertirlo a entero

catDF_US = catDF_US.astype({'id':int})

## ELiminamos cualquier duplicado
catDF_US.drop_duplicates(keep='first',inplace=True)
catDF_US
```

```python
## Ahora, vamos a combinar ambos dataframes para obtener nuestro dataframe final

## Primero, cambiamos el tipo de valor de "category_id" del dataframe principal a "int32",
## esto para que coincida con el "id" de nuestro jsonDataFrame

df_US_Videos = df_US_Videos.astype({'category_id':int})

## Segundo, utilizamos la función de pandas "merge" para combinar ambos dataframes
## tambien, aprovechamos en eliminar ambas columnas "category_id" e "id", solo necesitamos la descripción
df_US_Videos = pd.merge(df_US_Videos,catDF_US, how='inner',left_on="category_id",right_on="id").drop(['id','category_id'],axis=1)
df_US_Videos
```

```python
# Ahora debemos convertir las variables "trending_date" y "publish_date" al mismo formato (yyyy-mm-dd)

df_US_Videos['trending_date'] = pd.to_datetime(df_US_Videos['trending_date'], format="%Y-%m-%d")
df_US_Videos['publish_time'] = pd.to_datetime(df_US_Videos['publish_time'])
df_US_Videos
```

```python
# Antes de comenzar a análizar si existen valores nulos u outliers
# Debemos formatear la data
# Revisamos una vez más el dtype de nuestro dataframe
df_US_Videos.info()
```

```python
df_US_Videos['video_id'] = df_US_Videos['video_id'].astype('string')
df_US_Videos['title'] = df_US_Videos['title'].astype('string')
df_US_Videos['channel_title'] = df_US_Videos['channel_title'].astype('string')
df_US_Videos['tags'] = df_US_Videos['tags'].astype('string')
df_US_Videos['thumbnail_link'] = df_US_Videos['thumbnail_link'].astype('string')
df_US_Videos['description'] = df_US_Videos['description'].astype('string')
df_US_Videos['state'] = df_US_Videos['state'].astype('string')
df_US_Videos['snippet.title'] = df_US_Videos['snippet.title'].astype('string')
df_US_Videos.info()
```

```python
# Revisamos si existen datos nulls

print(f'{df_US_Videos.isnull().sum().sum()} datos nulls\nObservando la descripción general, podemos deducir que proviene de la columna "description"')
```

```python
'''
Si bien es cierto son pocos videos los que no presentan tanto likes como dislikes
esto no implica que esten vacios, ya que tomo estos valores como valores atípicos
que implican popularidad o despopularidad en la plataforma Youtube
por lo tanto no creo que deban ser eliminado

'''
```

```python
#Entonces, habiendo revisado el dataset en busca de datos incompletos, solo hemos dado con esos 570 filas que presentan
# en la variable 'description' datos vacios
# Como 570 no representa ni el 5% del total de datos, entonces se opta por eliminar las filas que presenten datos vacios
# 40949 

df_US_Videos.dropna(subset = ["description"], inplace=True)
```
